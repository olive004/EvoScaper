```python
def get_rna_links(url: str, base_url: str):
    for a in soup.find_all('a'):
        c = a.get('class')
        if (c is not None) and ('GENERIC-RNA' in c):
            l = base_url + a.get('href')
            print(l)

get_rna_links(url, base_url=base_url)
```
